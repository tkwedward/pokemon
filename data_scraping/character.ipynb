```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import pandas as pd
import numpy as np
```

```python
html = open('/Users/mac/Desktop/projects/pokemon/character-0.html').read()
soup = BeautifulSoup(html, 'html.parser')
```

```python
table = soup.find('table', attrs={'class':'character'})
```

```python
table
```

```python
# 找第三代的特性
def get_each_line(x):
        columns = x.find_all('td')
        cn_name=  x.find_all('td')[0].text.replace('\n','')
        eng_name=  x.find_all('td')[1].text.replace('\n','')
        jp_name=  x.find_all('td')[2].text.replace('\n','')
        atk=  x.find_all('td')[3].text.replace('\n','')
        df=  x.find_all('td')[4].text.replace('\n','')
        spAtk=  x.find_all('td')[5].text.replace('\n','')
        spDf=  x.find_all('td')[6].text.replace('\n','')
        spd=  x.find_all('td')[7].text.replace('\n','')
        
        return(list([cn_name, eng_name, jp_name, atk, df,spAtk, spDf, spd]))

def data_clean(gen):
    row = gen.find_all('tr')[1:]
    return list(map(get_each_line, row))
    
```

```python
all_list=data_clean(table)
```

```python
columns = table.find_all('td')
```

```python
all_list
```

```python
all_list_df = pd.DataFrame(all_list, columns=['中文', '英文', '日文', '攻擊', '防禦', '特攻', '特防', '速度'])
```

```python
all_list_df
```

```python
def replace_arrows(x):
    if x =="↑":
        return 1.1
    elif x =="↓":
        return 0.9
    elif x == "":
        return 1.0
    else:
        return x

# all_list_df.apply(replace_arrows, axis=1)
all_list_df = all_list_df.agg({'中文': replace_arrows, '英文': replace_arrows, '日文': replace_arrows, '攻擊': replace_arrows,'防禦': replace_arrows, '特攻': replace_arrows, '特防': replace_arrows, '速度': replace_arrows})
all_list_df
```
